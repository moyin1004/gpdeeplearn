```python
from transformers import AutoTokenizer
import transformers

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer: transformers.TokenizersBackend = AutoTokenizer.from_pretrained(checkpoint)
```

```python
raw_inputs = [
    "I've been waiting for a Hugging Face course my whole life.",
    "I hate this so much!"
]
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt")
inputs
```

```python
tokenizer.decode([  101,  1045,  1005,  2310,  2042,  3403,  2005,  1037, 17662,  2227,
          2607,  2026,  2878,  2166,  1012,   102])
```

```python
from transformers import AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(**inputs)
print(outputs.logits.shape)
```

padding 占位符

```python
sequence1_ids = [[200, 200, 200]]
sequence2_ids = [[200, 200]]

batched_ids = [
    [200, 200, 200],
    [200, 200, tokenizer.pad_token_id]
]

print(model(torch.tensor(sequence1_ids)).logits)
print(model(torch.tensor(sequence2_ids)).logits)

print(model(torch.tensor(batched_ids)).logits)
```

attention_mask

```python
attention_mask = [
    [1, 1, 1],
    [1, 1, 0] # 指定第二个序列的最后一个token不参与计算
]
outputs = model(torch.tensor(batched_ids), attention_mask=torch.tensor(attention_mask))
print(outputs.logits)
```

```python
sequences = ["I've been waiting for a Hugging Face course my whole life.", 
             "I hate this so much!",
             "hello world"]
```

```python
model_inputs = tokenizer(sequences, padding="longest")
model_inputs
```

```python
model_inputs = tokenizer(sequences, padding="max_length")
len(model_inputs.input_ids[0]) # BERT 默认长度最大为512
```

```python
model_inputs = tokenizer(sequences, padding=True, return_tensors="pt")
model_inputs
```
